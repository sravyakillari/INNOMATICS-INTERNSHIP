```python
#Problem Statement:
#Analyze student results.
marks = [45, 78, 90, 33, 60]
#a student will pass if he/she gets >=50
pass_count=0
fail_count=0
for i in marks:
    if i >=50:
        pass_count+=1#counts the no.of students passed
    else:
        fail_count+=1#counts the no.of students failed
print("Total No.of Students passed are: ",pass_count)
print("Total No.of Students failed are: ",fail_count)
```

```python
#Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]
a=[]
for i in names:
    a.append(i.strip().lower())#strip() removes extra spaces and lower() transforms to lowercase
print(a)
```

```python
#Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    print(i,'length is: ',len(i))#finds the length
    if len(i)>10:
        print(i,"message length is longer than 10 characters i.e, ", len(i))
```

```python
#Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i=='ERROR':
        count+=1#count increments when i matches with error
print('The total ERROR count is: ',count)
```

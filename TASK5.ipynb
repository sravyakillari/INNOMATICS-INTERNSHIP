```python
# Create a function to analyze engagement on a social media post.
def engagement(likes):
    total=0
    for like in likes:
        total+=like
#condition logic to detrmine post status
    if total>=1000:
        status="Viral Post"
    else:
        status="Normal Engagement"
    print("Total Likes:",total)
    print("Post STatus:",status)
likes=[120,300,250,380]
engagement(likes)
```

```python
# Create a function that analyzes seasonal rainfall data.
def rainfallcheck(data,required_level):
    total=0
    for rain in data:
        total+=rain
    #calculate average rainfall
    avg=total/len(data)
    if avg>=required_level:
        status="Adequate Rainfall"
    else:
        status="Indequate Rainfall"
    print("Average Rainfall:",round(avg,2))
    print("Rainfall Status:",status)
rainfall_data=[80,65,70,73]
required_level=70
rainfallcheck(rainfall_data,required_level)
```

```python
# Create a function to detect duplicate usernames in a system.
def detectdup(usernames):
    unique=set(usernames)
    if len(unique)<len(usernames):#condition to check whether duplicates are found or not
        status="Duplicate Accounts Found: Yes"
    else:
        status="Duplicate Accounts Found: No"
    print(status)
usernames=["sharanya","sravya","killari",'sharanya']
detectdup(usernames)
```

```python
# Create a function that filters crop prices above a premium threshold.
def premium(prices,threshold=2000):
    premium=[]
    for price in prices:
        if price>threshold:
            premium.append(price)
    print("Premium Crops:",premium)
prices=[1500,2500,1800,3200,2000]
premium(prices)
```

```python
# Create a function to evaluate application health based on error count.
def application_health(error):
    if error==0:
        status="Healthy"
    elif error<=5:
        status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",error)
    print("Application Status:",status)
application_health(7)
```

```python
# Create a function to calculate attendance percentage and check exam eligibility.
def attendance(records,total):
    attended=0
    for record in records:
        if record=="Present":
            attended+=1
    perc=(attended/total)*100
    if perc>=75:
        status="Eligible"
    else:
        status="Not Eligible"
    print("Attendance percentage:",round(perc,2))
    print("Exam Eligibility:",status)
records=["Present","Absent","Present","Present","Present"]
total=len(records)
attendance(records,total)
```

```python
# function to manage parking
def parking(capacity,vehicle_logs):
    parked=0
    peak_usage=0
    for log in vehicle_logs:
        if log=="IN":
            parked+=1
        elif log=="OUT" and parked>0:
            parked-=1
        if parked>peak_usage:
            peak_usage=parked
    if parked>capacity:
        status="Capacity Exceeded"
    else:
        status="Available"
    print("Currently Parked Vehicles:",parked)
    print("peak Usage:",peak_usage)
    print("Parking Status:",status)
capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
parking(capacity,vehicle_logs)
```

```python
# 2"online food delivery estimator
def delivery(distance,traffic,weather):
    basetime=distance*5
    #apply traffic delays
    if traffic=='Low':
        traffic_delay=0
    elif traffic=='Medium':
        traffic_delay=5
    else:
        traffic_delay=10
     #apply weather delays
    if weather=="Clear":
        weather_delay=0
    elif weather=="Rainy":
        weather_delay=5
    elif weather=="Stormy":
        weather_delay=15
    else:
        weather_delay=0
    eta=basetime+traffic_delay+weather_delay
    print("Estimated Delivery Time:",eta,"minutes")
    return eta
delivery(8,"High","Rainy")
```

```python
#3 : MOvie Theatre Seat A Occupancy Analyzer
def seat_occupancy(total,booked):
    booked_count=len(booked)
    #calculate occupancy percentage
    occupancy_percentage=(booked_count/total)*100
    if booked_count==total:
        status="Housefull"
    elif occupancy_percentage>=75:
        status="Almost Full"
    else:
        status="Available"
    suggestion="Open Additional Shows" if status=="Housefull" else "No Additional Shows Needed"
    print("Occupancy:",str(round(occupancy_percentage,2))+"%")
    print("Shows Status:",status)
    print("Suggestion:",suggestion)
total=200
booked=[1]*150
seat_occupancy(total,booked)
```

```python
#cloud server load classification system
def classify_server_load(cpu_readings):
    average_cpu=sum(cpu_readings)/len(cpu_readings)
    if average_cpu<50:
        status="Normal"
    elif 50<=average_cpu<80:
        status="Warning"
    else:
        status="Critical"
    print(f"Average CPU Load:{average_cpu:.2f}%")
    print(f"Server Status:{status}")
cpu_readings=[45,60,70,85,90]
classify_server_load(cpu_readings)
```

```python
#Smart Classroom Resource Usage Monitor
def monitor_resource_usage(resource_usage):
    thresholds={
        "Projector":5,
        "AC":8,
        "Lights":6
    }
    overused=[resource for resource,hours in resource_usage.items() if hours>thresholds[resource]]
    energy_alert="Yes" if overused else "No"
    if overused:
        print(f"Overused Resources:{','.join(overused)}")
    else:
        print("Overused Resources:None")
    print(f"Energy Alert:{energy_alert}")
resource_usage={
    "Projector":6,
    "AC":9,
    "Lights":4
}
monitor_resource_usage(resource_usage)
```

```python
# 1.smart light controller
class smartlight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"
    def turnon(self):
        self.status='ON'
    def turnoff(self):
        self.status="OFF"
    def display(self):
        print(f"{self.name} is {self.status}")
light=smartlight("Bedroom Light")
light.turnon()
light.display()
```

```python
#2.Employee ID card system
class Employee:
    def __init__(self,name,empid,dept):
        self.name=name
        self.empid=empid
        self.dept=dept
    def display(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.empid)
        print("Department:",self.dept)
emp=Employee("Rahul","EMP102","IT")
emp.display()
```

```python
#Mobile Contact Record
class Contact:
    def __init__(self,name,phone):
        self.phone=phone
        self.name=name
    def display(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.phone)
contact=Contact("Anitha",'9876543210')
contact.display()
```

```python
#4. product price tag generator
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print("Product:",self.name)
        print("Price: ₹",self.price)
p1=Product("Headphones",2499)
p1.display()
        
```

```python
#5. Movie Rating Display
class movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display(self):
        print("Movie:",self.name)
        print("Rating:",self.rating,'/ 5')
m1=movie("Inception",4.8)
m1.display()
```

```python
#6.Delivery address manager
class Delivery:
    def __init__(self,customer,address):
        self.customer=customer
        self.address=address
    def display(self):
        print("Delivery Details")
        print("Customer:",self.customer)
        print("Address:",self.address)
d1=Delivery("Suman","Hyderabad")
d1.display()
```

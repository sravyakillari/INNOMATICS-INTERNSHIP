```python
# Given a sentence, identify all unique words and display:
# The count of unique words
# The unique words themselves
sentence=input("Enter sentence: ")#for taking input
words=sentence.split()#the sentence will be splitted into words
unique=set(words)#convert to set to remove the duplicates
print("Unique words count:",len(unique))
print("Unique words:",unique)
```

```python
# Store employee names and their salaries, and determine which employee has the highest salary.
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}#input dictionary
highest=max(employees,key=employees.get)#finding employee with highest salary
print("Highest Salary:",highest,'-',employees[highest])
```

```python
# Find the maximum and minimum values from a list of numbers.
nums=[45,22,89,10,66]
#initializing min and max to first element in list
max1=nums[0]
min1=nums[0]
for num in nums:
    if num>max1:#if num greater than max1 then max1 will updated as num
        max1=num
    if num<min1:#same as max1 if it is lesser then min1 will updated as num
        min1=num
print("List:",nums)
print("Max:",max1)
print("Min:",min1)
```

```python
# Count how many products have a price greater than 1000.
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:#for loop through the list
    if i>1000:
        count+=1#if i is greater than 1000 then count will increase
print("Products above 1000:",count)
```

```python
# Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).
attendance = ["P", "P", "A", "P", "P"]
present=attendance.count("P")
percentage=(present/len(attendance))*100#calculate percentage
print("Attendance Percentage:",percentage)
```

```python
# A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
high=max(employees.values())#finding highest score
top=[name for name,score in employees.items() if score==high]
print(f'Top Performers Eligible for Bonus: {','.join(top)} (Score: {high})')
```

```python
# An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.
query="Buy mobile phone buy phone online"
words=query.lower().split()#convert to lower case
result={}
for word in words:
    result[word]=result.get(word,0)+1
result={word: count for word, count in result.items() if count>1}
print(result)
```

```python
# A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid=[(i,val) for i,val in enumerate(sensor_readings) if val%2==0]
print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
# A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domains=[email.split("@")[1] for email in emails]
#count frequency
count={}
for d in domains:
    count[d]=count.get(d,0)+1
total= len(emails)
for domain,count1 in count.items():
    percentage=(count1/total)*100
    print(f'{domain}:{int(percentage)}%')
```

```python
# A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.
sales=[1200,1500,900,2200,1400,3000]
avg=sum(sales)/len(sales)#average sales
for day,value in enumerate(sales,start=1):
    if value>=avg*1.3:
        print(f'Day {day}: {value}')
```

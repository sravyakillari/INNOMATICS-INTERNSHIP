```python
# Create a function to validate a mobile recharge request.
def validate(amount):
    validplans=[199,299,399,599]
    if amount<50:
        return False, "recharge amount must be at least 150"
    elif amount not in validplans:
        return False, f"Invalid plan. choose from {validplans}"
    else:
        return True,"Recharge successful"
while True:
    try:
        useramount=int(input("enter recharge amount:"))
        isvalid,message=validate(useramount)
        print(message)
        if isvalid:
            break
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        
```

```python

# Given a dictionary of products and stock quantities:
# Print Reorder Alert if stock < 15
# Print Stock OK otherwise
inventory={
    "Banana":10,
    "Apple":25,
    "Orange":14,
    "mango":18
}
def check(stock):
    for product,quantity in stock.items():
        if quantity<15:
            print(f"{product}:Reorder Alert")
        else:
            print(f"{product}: Stock ok")
check(inventory)
```

```python
 # Student Result Processing System
def process(marks):
    total=0
    for mark in marks:
        total+=mark
    average=total/len(marks)
    if average>=50:
        return "Pass"
        
    else:
        return "Fail"
marks=[60,45,70,55]
result=process(marks)
print("Average:",sum(marks)/len(marks))
print("Result:",result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate(distance,peakhr=False):
    basefare=50
    perkm=12
    fare=basefare+(distance*perkm)
    if peakhr:
        fare*=1.25
    return fare
while True:
    try:
        km=float(input("Enter distance travlled in(in km): "))
        peak=input("Is it peak hour? (yes/no): ").strip().lower()
        peakhr=True if peak=="yes" else False
        totalfare=calculate(km,peakhr)
        print(f"Total Fare: {totalfare: .2f}")
        retry=input("Do you want to calculate again? (yes/no): ").strip().lower()
        if retry!="yes":
            print("Thank you for using cab fare estimator!")
            break
    except ValueError:
        print("Invalid input. Please enter numeric values for distance.")
```
